````python
# llm-jp/llm-jp-3-13bã‚’4bité‡å­åŒ–ã®qLoRAè¨­å®šã§ãƒ­ãƒ¼ãƒ‰ã€‚

from unsloth import FastLanguageModel
import torch
max_seq_length = 512 # unslothã§ã¯RoPEã‚’ã‚µãƒãƒ¼ãƒˆã—ã¦ã„ã‚‹ã®ã§ã‚³ãƒ³ãƒ†ã‚­ã‚¹ãƒˆé•·ã¯è‡ªç”±ã«è¨­å®šå¯èƒ½
dtype = None # Noneã«ã—ã¦ãŠã‘ã°è‡ªå‹•ã§è¨­å®š
load_in_4bit = True # ä»Šå›ã¯13Bãƒ¢ãƒ‡ãƒ«ã‚’æ‰±ã†ãŸã‚True

model_id = "llm-jp/llm-jp-3-13b"
new_model_id = "llm-jp-3-13b-it" #Fine-Tuningã—ãŸãƒ¢ãƒ‡ãƒ«ã«ã¤ã‘ãŸã„åå‰ã€it: Instruction Tuning
# FastLanguageModel ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚’ä½œæˆ
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_id,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
    trust_remote_code=True,
)

# SFTç”¨ã®ãƒ¢ãƒ‡ãƒ«ã‚’ç”¨æ„
model = FastLanguageModel.get_peft_model(
    model,
    r = 32,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 32,
    lora_dropout = 0.05,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
    max_seq_length = max_seq_length,
)
````

````python
# å­¦ç¿’ã«ç”¨ã„ã‚‹ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®æŒ‡å®š
# ä»Šå›ã¯LLM-jp ã®å…¬é–‹ã—ã¦ã„ã‚‹ Ichikara Instruction ã‚’ä½¿ã„ã¾ã™ã€‚ãƒ‡ãƒ¼ã‚¿ã«ã‚¢ã‚¯ã‚»ã‚¹ã™ã‚‹ãŸã‚ã«ã¯ç”³è«‹ãŒå¿…è¦ã§ã™ã®ã§ã€ä½¿ã„ãŸã„æ–¹ã®ã¿ç”³è«‹ã‚’ã—ã¦ãã ã•ã„ã€‚
# Ichikara Instruciton ã‚’ Hugging Face Hub ã«ã¦å…¬é–‹ã™ã‚‹ã“ã¨ã¯ãŠæ§ãˆãã ã•ã„ã€‚
# ã¾ãŸã€CC-BY-NC-SAã§ã™ã®ã§ãƒ¢ãƒ‡ãƒ«ã¯ãƒ©ã‚¤ã‚»ãƒ³ã‚¹ã‚’ç¶™æ‰¿ã™ã‚‹å‰æã§ãŠä½¿ã„ãã ã•ã„ã€‚

# ä¸‹è¨˜ã®ãƒªãƒ³ã‚¯ã‹ã‚‰ç”³è«‹ã‚’çµ‚ãˆãŸå…ˆã« Google Drive ãŒã‚ã‚Šã€Distribution20241221_all ã¨ã„ã†ãƒ•ã‚©ãƒ«ãƒ€ã”ã¨ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„ã€‚
# ä»Šå›ã¯ã€Œichikara-instruction-003-001-1.jsonã€ã‚’ä½¿ã„ã¾ã™ã€‚å¿…è¦ã§ã‚ã‚Œã°å±•é–‹ï¼ˆ!unzip ãªã©ï¼‰ã—ã€ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®ãƒ‘ã‚¹ã‚’é©åˆ‡ã«æŒ‡å®šã—ã¦ãã ã•ã„ã€‚
# omnicampusã®é–‹ç™ºç’°å¢ƒã§ã¯å–å¾—ã—ãŸãƒ‡ãƒ¼ã‚¿ã‚’å·¦å´ã«ãƒ‰ãƒ©ãƒƒã‚°ã‚¢ãƒ³ãƒ‰ãƒ‰ãƒ­ãƒƒãƒ—ã—ã¦ãŠä½¿ã„ãã ã•ã„ã€‚
# Google Colab ã®å ´åˆã‚‚å·¦ã®ã‚µã‚¤ãƒ‰ãƒãƒ¼ã‚ˆã‚Šãƒ‰ãƒ©ãƒƒã‚°&ãƒ‰ãƒ­ãƒƒãƒ—ã§ã‚¢ãƒƒãƒ—ãƒ‡ãƒ¼ãƒˆã—ã¦ãã ã•ã„ã€‚

# https://liat-aip.sakura.ne.jp/wp/llmã®ãŸã‚ã®æ—¥æœ¬èªã‚¤ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚·ãƒ§ãƒ³ãƒ‡ãƒ¼ã‚¿ä½œæˆ/llmã®ãŸã‚ã®æ—¥æœ¬èªã‚¤ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚·ãƒ§ãƒ³ãƒ‡ãƒ¼ã‚¿-å…¬é–‹/
# é–¢æ ¹è¡, å®‰è—¤ã¾ã‚„, å¾Œè—¤ç¾çŸ¥å­, éˆ´æœ¨ä¹…ç¾, æ²³åŸå¤§è¼”, äº•ä¹‹ä¸Šç›´ä¹Ÿ, ä¹¾å¥å¤ªéƒ. ichikara-instruction: LLMã®ãŸã‚ã®æ—¥æœ¬èªã‚¤ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚·ãƒ§ãƒ³ãƒ‡ãƒ¼ã‚¿ã®æ§‹ç¯‰. è¨€èªå‡¦ç†å­¦ä¼šç¬¬30å›å¹´æ¬¡å¤§ä¼š(2024)

from datasets import load_dataset

dataset = load_dataset("json", data_files="/content/drive/MyDrive/ichikara-instruction-003-001-1.json")
# ãƒ‘ã‚¹ã®æŒ‡å®šã«ã”æ³¨æ„ãã ã•ã„ã€‚ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã—ãŸãƒ•ã‚¡ã‚¤ãƒ«ã‚’å³ã‚¯ãƒªãƒƒã‚¯ã—ã€ã€Œãƒ‘ã‚¹ã‚’ã‚³ãƒ”ãƒ¼ã€ã‚’ã‚¯ãƒªãƒƒã‚¯ã€ä¸Šè¨˜ã® data_files ã¨åˆè‡´ã—ã¦ã„ã‚‹ã“ã¨ã‚’ã”ç¢ºèªãã ã•ã„ã€‚Omnicampus ã®ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªæ§‹é€ ã¨ã¯ç•°ãªã‚‹ã‹ã‚‚ã—ã‚Œã¾ã›ã‚“ã€‚
````

````python
# å­¦ç¿’æ™‚ã®ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆãƒ•ã‚©ãƒ¼ãƒãƒƒãƒˆã®å®šç¾©
prompt = """### æŒ‡ç¤º
{}
### å›ç­”
{}"""


"""
formatting_prompts_func: å„ãƒ‡ãƒ¼ã‚¿ã‚’ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã«åˆã‚ã›ãŸå½¢å¼ã«åˆã‚ã›ã‚‹
"""
EOS_TOKEN = tokenizer.eos_token # ãƒˆãƒ¼ã‚¯ãƒŠã‚¤ã‚¶ãƒ¼ã®EOSãƒˆãƒ¼ã‚¯ãƒ³ï¼ˆæ–‡æœ«ãƒˆãƒ¼ã‚¯ãƒ³ï¼‰
def formatting_prompts_func(examples):
    input = examples["text"] # å…¥åŠ›ãƒ‡ãƒ¼ã‚¿
    output = examples["output"] # å‡ºåŠ›ãƒ‡ãƒ¼ã‚¿
    text = prompt.format(input, output) + EOS_TOKEN # ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã®ä½œæˆ
    return { "formatted_text" : text, } # æ–°ã—ã„ãƒ•ã‚£ãƒ¼ãƒ«ãƒ‰ "formatted_text" ã‚’è¿”ã™
pass

# # å„ãƒ‡ãƒ¼ã‚¿ã«ãƒ•ã‚©ãƒ¼ãƒãƒƒãƒˆã‚’é©ç”¨
dataset = dataset.map(
    formatting_prompts_func,
    num_proc= 4, # ä¸¦åˆ—å‡¦ç†æ•°ã‚’æŒ‡å®š
)

dataset
````

````python
"""
training_arguments: å­¦ç¿’ã®è¨­å®š

  - output_dir:
      -ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°å¾Œã®ãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜ã™ã‚‹ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒª

  - per_device_train_batch_size:
      - ãƒ‡ãƒã‚¤ã‚¹ã”ã¨ã®ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ãƒãƒƒãƒã‚µã‚¤ã‚º

  - per_device_eval_batch_size:
      - ãƒ‡ãƒã‚¤ã‚¹ã”ã¨ã®è©•ä¾¡ãƒãƒƒãƒã‚µã‚¤ã‚º

  - gradient_accumulation_steps:
      - å‹¾é…ã‚’æ›´æ–°ã™ã‚‹å‰ã«ã‚¹ãƒ†ãƒƒãƒ—ã‚’ç©ã¿é‡ã­ã‚‹å›æ•°

  - optim:
      - ã‚ªãƒ—ãƒ†ã‚£ãƒã‚¤ã‚¶ã®è¨­å®š

  - num_train_epochs:
      - ã‚¨ãƒãƒƒã‚¯æ•°

  - eval_strategy:
      - è©•ä¾¡ã®æˆ¦ç•¥ ("no"/"steps"/"epoch")

  - eval_steps:
      - eval_strategyãŒ"steps"ã®ã¨ãã€è©•ä¾¡ã‚’è¡Œã†stepé–“éš”

  - logging_strategy:
      - ãƒ­ã‚°è¨˜éŒ²ã®æˆ¦ç•¥

  - logging_steps:
      - ãƒ­ã‚°ã‚’å‡ºåŠ›ã™ã‚‹ã‚¹ãƒ†ãƒƒãƒ—é–“éš”

  - warmup_steps:
      - å­¦ç¿’ç‡ã®ã‚¦ã‚©ãƒ¼ãƒ ã‚¢ãƒƒãƒ—ã‚¹ãƒ†ãƒƒãƒ—æ•°

  - save_steps:
      - ãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜ã™ã‚‹ã‚¹ãƒ†ãƒƒãƒ—é–“éš”

  - save_total_limit:
      - ä¿å­˜ã—ã¦ãŠãcheckpointã®æ•°

  - max_steps:
      - ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ã®æœ€å¤§ã‚¹ãƒ†ãƒƒãƒ—æ•°

  - learning_rate:
      - å­¦ç¿’ç‡

  - fp16:
      - 16bitæµ®å‹•å°æ•°ç‚¹ã®ä½¿ç”¨è¨­å®šï¼ˆç¬¬8å›æ¼”ç¿’ã‚’å‚è€ƒã«ã™ã‚‹ã¨è‰¯ã„ã§ã™ï¼‰

  - bf16:
      - BFloat16ã®ä½¿ç”¨è¨­å®š

  - group_by_length:
      -  å…¥åŠ›ã‚·ãƒ¼ã‚±ãƒ³ã‚¹ã®é•·ã•ã«ã‚ˆã‚Šãƒãƒƒãƒã‚’ã‚°ãƒ«ãƒ¼ãƒ—åŒ– (ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ã®åŠ¹ç‡åŒ–)

  - report_to:
      - ãƒ­ã‚°ã®é€ä¿¡å…ˆ ("wandb"/"tensorboard"ãªã©)
"""
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset=dataset["train"],
    max_seq_length = max_seq_length,
    dataset_text_field="formatted_text",
    packing = False,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        num_train_epochs = 1,
        logging_steps = 10,
        warmup_steps = 10,
        save_steps=100,
        save_total_limit=2,
        max_steps=-1,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        group_by_length=True,
        seed = 3407,
        output_dir = "outputs",
        report_to = "none",
    ),
)
````

````python
# ELYZA-tasks-100-TVã®èª­ã¿è¾¼ã¿ã€‚äº‹å‰ã«ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„
# ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã®èª­ã¿è¾¼ã¿ã€‚
# omnicampusã®é–‹ç™ºç’°å¢ƒã§ã¯ã€å·¦ã«ã‚¿ã‚¹ã‚¯ã®jsonlã‚’ãƒ‰ãƒ©ãƒƒã‚°ã‚¢ãƒ³ãƒ‰ãƒ‰ãƒ­ãƒƒãƒ—ã—ã¦ã‹ã‚‰å®Ÿè¡Œã€‚
import json
datasets = []
with open("/content/drive/MyDrive/elyza-tasks-100-TV_0.jsonl", "r") as f:
    item = ""
    for line in f:
      line = line.strip()
      item += line
      if item.endswith("}"):
        datasets.append(json.loads(item))
        item = ""
````

````python
# å­¦ç¿’ã—ãŸãƒ¢ãƒ‡ãƒ«ã‚’ç”¨ã„ã¦ã‚¿ã‚¹ã‚¯ã‚’å®Ÿè¡Œ
from tqdm import tqdm

# æ¨è«–ã™ã‚‹ãŸã‚ã«ãƒ¢ãƒ‡ãƒ«ã®ãƒ¢ãƒ¼ãƒ‰ã‚’å¤‰æ›´
FastLanguageModel.for_inference(model)

results = []
for dt in tqdm(datasets):
  input = dt["input"]

  prompt = f"""### æŒ‡ç¤º\n{input}\n### å›ç­”\n"""

  inputs = tokenizer([prompt], return_tensors = "pt").to(model.device)

  outputs = model.generate(**inputs, max_new_tokens = 512, use_cache = True, do_sample=False, repetition_penalty=1.2)
  prediction = tokenizer.decode(outputs[0], skip_special_tokens=True).split('\n### å›ç­”')[-1]

  results.append({"task_id": dt["task_id"], "input": input, "output": prediction})
````

````python
# jsonlã§ä¿å­˜
with open(f"{new_model_id}_output.jsonl", 'w', encoding='utf-8') as f:
    for result in results:
        json.dump(result, f, ensure_ascii=False)
        f.write('\n')
````
